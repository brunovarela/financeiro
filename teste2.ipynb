```python
import yfinance as yf
import pandas as pd

# Defina a lista de tickers e o período de tempo
tickers = ["BBDC4.SA", "ITUB4.SA", "BRBI11.SA"]
wallet = {
    "Ticket": ["BBDC4.SA", "ITUB4.SA", "BRBI11.SA"],
    "Amount": [254, 108, 185]
}
start_date = "2024-08-12"
end_date = "2024-09-05"

# Baixe os dados históricos para vários tickers
stock_data = yf.download(tickers, start=start_date, end=end_date, group_by='ticker')

stock_data

```

```python
for ticker, amount in zip(wallet["Ticket"], wallet["Amount"]):
    # Cria a coluna 'Total' com None para o ticker
    stock_data[(ticker, 'Total')] = None
    stock_data[(ticker, 'Geral')] = None
    print(ticker)
    # Itera sobre cada linha do DataFrame
    for idx, row in enumerate(stock_data.itertuples()):
        #print(stock_data[(ticker, 'Close')][idx])
        stock_data[(ticker, 'Total')][idx] = stock_data[(ticker, 'Close')][idx]*amount





```
